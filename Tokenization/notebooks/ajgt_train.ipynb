```python
import glob
import json
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
data = []
path = 'ckpts/**/ajgt/accuracies.json'

for path in glob.glob(path):
    vocab_size = path.split('/')[1].split('-')[-1]
    dd = json.load(open(path, mode='r'))
    for tokenizer in dd:
        results = dd[tokenizer]
        tokenizer_name = tokenizer.replace(f'-{vocab_size}', '').replace('-ar', '')
        for val in results:
            data.append([int(vocab_size)]+[tokenizer_name]+[float(val)])
df = pd.DataFrame(data, columns = ['vocab_size', 'tokenizer', 'accuracy'])
g = sns.boxplot(x="vocab_size", y="accuracy", hue="tokenizer",
                 data=df, palette="Set3")
# g.set(ylim=(0, 1))
```
